2. 在資料清理的過程中，我們可以使用填補空值的方法避免遺失值帶來的錯誤。實務上會建議「連續數值的話補平均，離散數值的話補眾數」，請問為什麼？

(在這裡回答)
連續數值的話補平均: 保持常態分佈

離散數值的話補眾數: 保持離散的特性

為了保持資料原有分布的原則，當遇到缺失的連續數值資料時，使用平均值填補可以維持數據的中心趨勢，不會引入太大的偏差；對於離散數值，使用眾數（即出現次數最多的值）作為填補值，是因為這能夠保留數據的最常見特徵

平均值容易受到極值影響，所以在離散數值上不建議用平均，較建議用眾數，而在連續數值上，平均值較能保持數據的整體結構和特性。

整體來說：Missing Value 的處理策略以「盡量不要破壞原有的分佈」為原則



3. 承上題，也有人會建議可以利用統計內差/回歸或機器學習的方法進行空值填補，你覺得這是一種好方法嗎？為什麼？

(在這裡回答)

若是數據呈現數據時間序列或是線性關係利用統計內差會是一項好選擇；
但是數據呈現複雜的非線性關係利用機器學習來填補會較統計方法來的好

內插/回歸方法基於現有數據的統計特徵，對於數據的整體分佈有較好的保留性，填補的結果相對穩定，但當數據存在非線性關係時，統計內差的效果可能會受到限制。機器學習方法可以更好地捕捉數據之間的複雜關係，對於非線性關係的數據填補效果較好

4. 可以請你解釋一下「Label-Encoding」和「One-Hot-Encoding」分別用於什麼情境？

(在這裡回答)

label -> Encoding : 將類別資料轉換成數值順序，將該欄類別映射到整數，再有明顯程度大小上差異時可以使用，通常，類別轉換成數值形式會有大小之分。如果該欄位沒有程度上的差異(例如性別)，可能就比較不適合 LabelEncoder，會改用 OneHotEncoder。

One-Hot-Encoding : 將類別欄位轉換為數值欄位的方法，將每個類別值轉換為一個二進制向量，
用在無序的類別資料中。這可以防止模型錯誤地認為類別間存在數值上的大小關係，適用於大多數的機器學習算法。

    - Drop missing observations
    - Drop rows where all cells in that row is NA
    - Create a new column full of missing values
    - Fill in missing data with zeros
    - Fill in missing in preTestScore with the mean value of preTestScore
    - Fill in missing in postTestScore with each sex’s mean value of postTestScore
    - Select some rows but ignore the missing data points
    

```python
#sample code

import numpy as np
import pandas as pd
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'],
        'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'],
        'age': [42, np.nan, 36, 24, 73],
        'sex': ['m', np.nan, 'f', 'm', 'f'],
        'preTestScore': [4, np.nan, np.nan, 2, 3],
        'postTestScore': [25, np.nan, np.nan, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'sex', 'preTestScore', 'postTestScore'])

# 刪除缺失觀測值
# 刪除所有欄位皆為 NA 的列
# 新增一個充滿缺失值的新欄位
# 用零填補缺失資料
# 用 preTestScore 的平均值填補缺失的 preTestScore 資料
# 用每個性別 postTestScore 的平均值填補缺失的 postTestScore 資料
# 選擇某些列但忽略缺失資料點


df.isnull() #true is missing Value
```

```python
#刪除缺失觀測值 Drop missing observations
## 刪除所有含有缺失值的觀測值（列）
df.dropna()
#刪除含有 Missing Value 的觀測值（列）是一個常見的處理策略，但需要注意的是，這樣做可能會導致資料損失，因此需要根據具體情況進行決策。
```

```python
## 刪除所有欄位皆為 NA 的列
df.dropna(thresh=1)
#保留至少有n个非NaN数据的行/列（用axis=0或1表示）,thresh=1代表只要有一筆正常的數值非NaN就會保留，反之全部都不正常就會刪除該row。
```

```python
# 新增一個充滿缺失值的新欄位
df['missing values'] = np.nan
df
#新增一個全部為 NaN 的欄位，是因為 np.nan 代表 NaN（Not a Number），是用來表示缺失值的一種特殊值。
#當將 np.nan 指定給 DataFrame 的某個欄位時，Pandas 會將該欄位的所有值都設置為 NaN，因此你會看到新增的這個欄位全部都是 NaN。
```
