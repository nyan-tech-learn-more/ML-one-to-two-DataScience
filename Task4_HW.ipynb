2. 在資料清理的過程中，我們可以使用填補空值的方法避免遺失值帶來的錯誤。實務上會建議「連續數值的話補平均，離散數值的話補眾數」，請問為什麼？

(在這裡回答)
連續數值的話補平均: 保持常態分佈

離散數值的話補眾數: 保持離散的特性

為了保持資料原有分布的原則，當遇到缺失的連續數值資料時，使用平均值填補可以維持數據的中心趨勢，不會引入太大的偏差；對於離散數值，使用眾數（即出現次數最多的值）作為填補值，是因為這能夠保留數據的最常見特徵

平均值容易受到極值影響，所以在離散數值上不建議用平均，較建議用眾數，而在連續數值上，平均值較能保持數據的整體結構和特性。

整體來說：Missing Value 的處理策略以「盡量不要破壞原有的分佈」為原則

(在這裡回答)

若是數據呈現數據時間序列或是線性關係利用統計內差會是一項好選擇；
但是數據呈現複雜的非線性關係利用機器學習來填補會較統計方法來的好

內插/回歸方法基於現有數據的統計特徵，對於數據的整體分佈有較好的保留性，填補的結果相對穩定，但當數據存在非線性關係時，統計內差的效果可能會受到限制。機器學習方法可以更好地捕捉數據之間的複雜關係，對於非線性關係的數據填補效果較好

4. 可以請你解釋一下「Label-Encoding」和「One-Hot-Encoding」分別用於什麼情境？

(在這裡回答)

label -> Encoding : 將類別資料轉換成數值順序，將該欄類別映射到整數，再有明顯程度大小上差異時可以使用，通常，類別轉換成數值形式會有大小之分。如果該欄位沒有程度上的差異(例如性別)，可能就比較不適合 LabelEncoder，會改用 OneHotEncoder。

One-Hot-Encoding : 將類別欄位轉換為數值欄位的方法，將每個類別值轉換為一個二進制向量，
用在無序的類別資料中。這可以防止模型錯誤地認為類別間存在數值上的大小關係，適用於大多數的機器學習算法。

```python
import numpy as np
import pandas as pd
url = 'https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'
df = pd.read_csv(url)
df
```

```python
Col_sum = df.isnull().sum() # true is missing Value
print(Col_sum) #加總所有missing value
print(Col_sum.max()) #最大值
```

```python
#如何用平均值替換「多個"數字"型態」列(collumn)的缺失值？
numeric_columns = df.select_dtypes(include=['float','int']).columns.tolist()
numeric_columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
df.isnull().sum()
```

```python

# 計算不同製造商的平均價格。
avg_price_ma = df.groupby('Manufacturer')['Price'].mean()
avg_price_ma
```

```python
# 如何根據製造商的平均值替換「價格」列的缺失值？
# df['Price'] = df.groupby('Manufacturer')['Price'].transform(lambda x: x.fillna(x.mean()))
# df.isnull().sum()


# Create a dictionary to store mean prices for each manufacturer
manufacturer_mean_prices = {}

# Iterate through each manufacturer group
for manufacturer in df['Manufacturer'].unique():
    # Get the price data for the current manufacturer
    manufacturer_prices = df[df['Manufacturer'] == manufacturer]['Price']

    # Calculate the mean price for the current manufacturer
    mean_price = manufacturer_prices.mean()

    # Store the mean price in the dictionary
    manufacturer_mean_prices[manufacturer] = mean_price

# Replace missing values in 'Price' using the manufacturer's mean
for index, row in df.iterrows():
    manufacturer = row['Manufacturer']
    price = row['Price']

    if pd.isna(price):
        # If the price is missing, use the mean price for the manufacturer
        df.loc[index, 'Price'] = manufacturer_mean_prices[manufacturer]

# Check for remaining missing values
print("Missing values after imputation:")
print(df.isnull().sum())
```

```python
import numpy as np
import pandas as pd
url = 'https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'
df = pd.read_csv(url)
df

Col_sum = df.isnull().sum() # true is missing Value
print(Col_sum) #加總所有missing value
print(Col_sum.max()) #最大值

#如何用平均值替換「多個"數字"型態」列(collumn)的缺失值？
numeric_columns = df.select_dtypes(include=['float','int']).columns.tolist()
numeric_columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
df.isnull().sum()


# 計算不同製造商的平均價格。
avg_price_ma = df.groupby('Manufacturer')['Price'].mean()
avg_price_ma

# 如何根據製造商的平均值替換「價格」列的缺失值？
# df['Price'] = df.groupby('Manufacturer')['Price'].transform(lambda x: x.fillna(x.mean()))
# df.isnull().sum()


# Create a dictionary to store mean prices for each manufacturer
manufacturer_mean_prices = {}

# Iterate through each manufacturer group
for manufacturer in df['Manufacturer'].unique():
    # Get the price data for the current manufacturer
    manufacturer_prices = df[df['Manufacturer'] == manufacturer]['Price']

    # Calculate the mean price for the current manufacturer
    mean_price = manufacturer_prices.mean()

    # Store the mean price in the dictionary
    manufacturer_mean_prices[manufacturer] = mean_price

# Replace missing values in 'Price' using the manufacturer's mean
for index, row in df.iterrows():
    manufacturer = row['Manufacturer']
    price = row['Price']

    if pd.isna(price):
        # If the price is missing, use the mean price for the manufacturer
        df.loc[index, 'Price'] = manufacturer_mean_prices[manufacturer]

# Check for remaining missing values
print("Missing values after imputation:")
print(df.isnull().sum())
```
