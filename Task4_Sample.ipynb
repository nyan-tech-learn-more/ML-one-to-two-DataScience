資料過濾

```python
import pandas as pd

data = pd.DataFrame({
    'col1': [5, 12, 8, 15],
    'col2': [16, 9, 18, 4],
    'col3': [11, 3, 7, 20]
})
#選擇col1 大於10的資料
subset1 = data.query('col1 > 10')
print(subset1)

#選擇col1 大於10且col2 < 20
subset2 = data[(data['col1'] > 10) & (data['col2'] < 20)]

print(subset2)
```

資料排序

```python
data = pd.DataFrame({
    'col1': [5, 12, 8, 15],
    'col2': [16, 9, 18, 4],
    'col3': [11, 3, 7, 20]
})

#按照col1排序
col1 = data.sort_values('col1')
print(col1) #data的col1被排序，其他相同位置的也會一起被排

# 按照 col1 和 col2 排序
col2 = data.sort_values(['col1', 'col2'])
print(col2)
```

資料去重複

```python
import pandas as pd

data = pd.DataFrame({
    'col1': [5, 12, 5, 8, 12],
    'col2': [16, 9, 16, 18, 9],
    'col3': [11, 3, 11, 7, 3]
})

#去除重複資料
dup = data.drop_duplicates()
print(dup)

#這個 dataframe 包含了 5 個列和 3 個欄，每一個列都有 col1、col2 和 col3 這三個欄位的數據。由於前兩列和後兩列的 col1 和 col2 都是一樣的，因此如果使用 drop_duplicates() 函數，這些重複的資料就會被刪除。
```

缺失值處理策略與方法

利用 Pandas 處理 Missing Value 的函式

```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'col1': [5, 12, 8, np.nan],
    'col2': [16, 9, np.nan, 4],
    'col3': [11, 3, 7, 20]
})


data.isnull() #檢查缺失數值
#。使用 isnull() 函數可以檢查出哪些資料是缺失值，這些缺失值會被標記為 True，而其他資料則會被標記為 False。
```

```python
#`notnull()` 方法則與 isnull 方法相反，返回 True 表示非 Missing Value：
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'col1': [5, 12, 8, np.nan],
    'col2': [16, 9, np.nan, 4],
    'col3': [11, 3, 7, 20]
})

# 檢查非缺失值
data.notnull() #非缺失notnull數值
```

```python
#`fillna()` 方法可以將 Missing Value 填充為指定的值：
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'col1': [5, 12, 8, np.nan],
    'col2': [16, 9, np.nan, 4],
    'col3': [11, 3, 7, 20]
})

data.fillna(0) #fill na 0 ->把NA填寫成0 (?)
#使用 fillna() 函數可以將缺失值填充為指定的數值，這個例子中是 0。
```

```python
#刪除含有 Missing Value 的觀測值（列）是一個常見的處理策略，但需要注意的是，這樣做可能會導致資料損失，因此需要根據具體情況進行決策。

import pandas as pd
import numpy as np

data = pd.DataFrame({
    'col1': [5, 12, np.nan, np.nan], #有缺資料
    'col2': [16, 9, np.nan, 4], #有缺資料
    'col3': [11, 3, 7, 20]
})

print(data)

#刪除所有含有缺失數值的觀測值(列)
data.dropna()
```

填充 Missing Value

```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'col1': [5, 12, 8, np.nan],
    'col2': [16, 9, np.nan, 4],
    'col3': [11, 3, 7, 20]
})
print(data)
print('-------')
data.fillna(0, inplace=True) #缺失都填0,取代原始資料
print(data) #輸出原始資料
```

```python
# 如果你想要修改原始 dataframe，必須將 inplace 參數設為 True。

# 使用平均值、中位數或眾數填充 Missing Value 是一種常見的方法。可以使用 Pandas 的 mean、median 或 mode 方法來計算平均值、中位數或眾數。

```

```python
import pandas as pd

data = pd.DataFrame({
    'col1': [5, 12, 8, np.nan],
    'col2': [16, 9, np.nan, 4],
    'col3': [11, 3, 7, 20]
})
print(data)
print('--------')
#將所有缺失值填充為平均值
data.fillna(data.mean())
```

```python
# 將所有缺失值填充為中位數
data.fillna(data.median())
```

```python
# 將所有缺失值填充為眾數（data.mode() 回傳結果是一個 df，要用 iloc 取得 series）
data.fillna(data.mode().iloc[0])

#眾數（英語：mode）指一組數據中出現次數最多的數據值。 例如{8，7，7，8，6，5，5，8，8，8}中，出現最多的是8，因此眾數是8，眾數可能是一個數（數據值），但也可能是多個數（數據值）。
```

④ 使用插值法填充 Missing Value

```python
#插值法是一種常見的填充 Missing Value 的方法，可以使用 Pandas 的 interpolate 方法實現。
#插值（英語：Interpolation），是一種通過已知的、離散的數據點，在範圍內推求新數據點的過程或方法。 求解科學和工程的問題時，通常有許多數據點藉由採樣、實驗等方法獲得，這些數據可能代表了有限個數值函數，其中自變量的值
import pandas as pd

data = pd.DataFrame({
    'col1': [5, 12, 8, np.nan],
    'col2': [16, 9, np.nan, 4],
    'col3': [11, 3, 7, 20]
})
print(data)
print('-----')
#使用插值法填充缺失值
data.interpolate()
```

```python
#⑤ 使用機器學習算法填充 Missing Value
# 機器學習算法可以用於填充 Missing Value，例如 K-Nearest Neighbor（KNN）算法。KNN 算法可以根據與缺失值最接近的 k 個樣本的值來填充缺失值。

import pandas as pd
from sklearn.impute import KNNImputer

data = pd.DataFrame({
    'col1': [5, 12, 8, np.nan],
    'col2': [16, 9, np.nan, 4],
    'col3': [11, 3, 7, 20]
})

print(data)
print('-----')

# 使用 KNN 算法填充缺失值
imputer = KNNImputer(n_neighbors=2)
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
data_imputed
```

類別欄位轉數值欄位

```python
#Pandas 提供的的欄位類型
#dtype

import pandas as pd

# 建立一個字典，包含各欄位的資料
data_dict = {'col1': [1, 2, 3, 4],
             'col2': ['A', 'B', 'C', 'D'],
             'col3': [0.1, 0.2, 0.3, 0.4]}

#使用字典建立一個DataFrame
data = pd.DataFrame(data_dict)

print(data.dtypes) ## 查詢每個欄位的類型

#這個範例程式碼會建立一個包含三個欄位的 dataframe，欄位名稱分別為 col1、col2 和 col3，分別代表整數、字串和浮點數類型的資料。
```

```python
#類別欄位轉數值欄位

import pandas as pd

# 創建一個包含類別欄位的 dataframe
df = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'category': ['A', 'B', 'C', 'A', 'B']
})
print(df)
```

```python
#Labelencoding Labelencoding 是一種將類別欄位轉換為數值欄位的方法，它可以將每個類別值映射為一個整數。Pandas 提供了 LabelEncoder 方法來實現 Labelencoding。
#將類別資料轉換成數值順序，將該欄類別映射到整數，不會新增欄位。像是衣服的size有分，假設某一份資料有 S、Ｍ、Ｌ、XL，依照 LabelEncoder 可以分成 0、1、2、3。而此時衣服大小有程度上的差異時，就會比較適合使用 LabelEncoder，讓機器學習該欄位(因子)的大小關係。
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['category'] = label_encoder.fit_transform(df['category']) #category欄位做labelencoding
print(df) #category 欄位被轉換為了 category_labelencoded 欄位，其中類別 A 被轉換為 0，類別 B 被轉換為 1，類別 C 被轉換為 2。
```

```python
#One-hot-encoding

#One-hot-encoding 是一種將類別欄位轉換為數值欄位的方法，它可以將每個類別值轉換為一個二進制向量。Pandas 提供了 get_dummies 方法來實現 One-hot-encoding。
```

```python
import pandas as pd

df = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'category': ['A', 'B', 'C', 'A', 'B']
})
print(df)
print('------')
# 使用 One-hot-encoding 將類別欄位轉換為數值欄位

df_onehotencoded = pd.get_dummies(df, columns=['category'])

print(df_onehotencoded)
df_onehotencoded
```
